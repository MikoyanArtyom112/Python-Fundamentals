````python
x = 6

while x > 4:
    x = x - 1
    print(x)
````

The loop printed out 4 because it continued executing until the value of `x` reached `4`, and then stopped. Let's go through it again:

1. **Initialization:** The loop starts with `x` set to `6`.
2. **Condition Check:** The condition `x > 4` is `True` because `6` is greater than `4`.
3. **First Iteration:** Inside the loop, `x` is decreased by `1`, so it becomes `5`, and `5` is printed.
4. **Second Iteration:** The loop goes back to the condition check. `x` is now `5`, and the condition `x > 4` is still `True`.
5. **Third Iteration:** `x` is decreased by `1` again, becoming `4`, and `4` is printed.
6. **Fourth Iteration:** Now, `x` is `4`. The condition `x > 4` is `False` because `4` is not greater than `4`.
7. **Loop Termination:** Since the condition is `False`, the loop stops executing.

So, the loop prints out `5` and `4` because those are the values of `x` that satisfy the condition `x > 4`. Once `x` becomes `4`, the loop stops, and no further values are printed.


In a `while` loop, the values are printed out as the loop iterates through each iteration. However, it only prints the values that meet the condition specified in the loop. In this case, the condition is `x > 4`, so the loop only prints values of `x` that are greater than `4`.

Each time the loop runs, it decreases the value of `x` by `1` until `x` is no longer greater than `4`. Once `x` reaches `4`, the loop stops, and no further values are printed because `4` does not meet the condition `x > 4`.

In contrast, in a `for` loop, you explicitly specify the sequence of values to iterate over, and it prints each value in that sequence, regardless of any condition. But in a `while` loop, the condition determines whether the loop continues or stops, and it only prints values that satisfy that condition.


### Exploring while Loops A Bit More

````python
# Example 1: Using a for loop to iterate over a sequence
print("Example 1: Using a for loop")
for i in range(5):  # Using range() to create a sequence from 0 to 4
    print(i)        # Printing the current value of i

# Example 2: Using a while loop to iterate until a condition is met
print("\nExample 2: Using a while loop")
j = 0               # Initialize j to 0
while j < 5:        # Loop until j is less than 5
    print(j)        # Printing the current value of j
    j += 1          # Incrementing j by 1 in each iteration

# Example 3: Using both for and while loops together
print("\nExample 3: Using both for and while loops")
for k in range(3):  # Using a for loop to iterate 3 times
    m = 0           # Initialize m to 0 for each iteration of the outer loop
    while m < 2:    # Nested while loop to iterate until m is less than 2
        print(k, m) # Printing the current values of k and m
        m += 1      # Incrementing m by 1 in each iteration of the inner loop
    # End of the inner while loop
# End of the outer for loop
````

### Nesting Loops in Loops

````python
i = 0
while i < 3:
    j =8
    while j <11:
        print(i,j)
        j = j + 1
    i = i + 1
````

The code consists of nested while loops, which means there is a while loop inside another while loop. Let's go through it step by step:

1. **Initialization:** The outer while loop starts with `i` set to `0`.

2. **Outer Loop Condition Check:** The condition `i < 3` is `True` because `0` is less than `3`.

3. **Outer Loop First Iteration:** Inside the outer loop, `j` is initialized to `8`. Then, the inner while loop starts.

4. **Inner Loop Condition Check:** The condition `j < 11` is `True` because `8` is less than `11`.

5. **Inner Loop Iterations:** Inside the inner loop, the `print()` function is called with the current values of `i` and `j`. In this case, `i` is `0` and `j` ranges from `8` to `10`, printing the pairs `(0, 8)`, `(0, 9)`, and `(0, 10)`.

6. **Inner Loop Termination:** After the inner loop completes its iterations, `j` is incremented to `11`, and the inner loop exits.

7. **Outer Loop Iteration:** After the inner loop, `i` is incremented by `1`, becoming `1`. Then, the outer loop goes back to check its condition.

8. **Outer Loop Condition Check (again):** The condition `i < 3` is still `True` because `1` is less than `3`.

9. **Outer Loop Second Iteration:** Inside the outer loop, `j` is again initialized to `8`, and the inner loop restarts.

10. **Inner Loop Iterations (again):** Inside the inner loop, the `print()` function is called again with the updated values of `i` and `j`. This time, it prints the pairs `(1, 8)`, `(1, 9)`, and `(1, 10)`.

11. **Inner Loop Termination (again):** After the inner loop completes its iterations again, `j` is incremented to `11`, and the inner loop exits again.

12. **Outer Loop Iteration (again):** After the second iteration of the inner loop, `i` is incremented by `1` again, becoming `2`, and the outer loop goes back to check its condition.

13. **Outer Loop Condition Check (last time):** The condition `i < 3` is still `True` because `2` is less than `3`.

14. **Outer Loop Third Iteration:** Inside the outer loop, `j` is initialized to `8` once more, and the inner loop starts again.

15. **Inner Loop Iterations (last time):** Inside the inner loop, the `print()` function is called again with the updated values of `i` and `j`. This time, it prints the pairs `(2, 8)`, `(2, 9)`, and `(2, 10)`.

16. **Inner Loop Termination (last time):** After the inner loop completes its iterations for the last time, `j` is incremented to `11`, and the inner loop exits for the final time.

17. **Outer Loop Termination:** After the outer loop completes its iterations, `i` is incremented to `3`, and the outer loop exits.

18. **End of Execution:** The code execution ends, and all the pairs `(i, j)` where `i` ranges from `0` to `2` and `j` ranges from `8` to `10` are printed.

### The Story of the Decision Maker (if) and the Repeater (while)

Imagine you're the captain of a ship sailing the vast ocean, and you have two key crew members: the Decision Maker and the Repeater.

The Decision Maker: if

**The Decision Maker** 
(the if statement) is like a crew member who makes single decisions based on specific conditions at a moment in time. When you approach an island, the Decision Maker looks through the telescope and decides if the island has a treasure. If there is treasure, you may choose to land; if not, you continue sailing. The Decision Maker doesn't care about past decisions or future possibilities beyond this one condition.


````python
# There's an island with a treasure map
treasure_map = True

if treasure_map:
    print("Land and search for the treasure!")
````

Here, the decision is simple: if there's a treasure map, you land; otherwise, you do nothing further about it.

**The Repeater: while**
The Repeater (the while loop) is like a crew member who repeats a specific action over and over until a certain condition changes. Imagine you're caught in a storm, and the Repeater's job is to keep bailing water out of the boat until the storm passes. They keep doing it — bail water, check the weather, and if the storm is still going, they continue bailing.


How to Choose Between if and while
- Use if when: You need to make a decision once, based on the current state of variables or inputs. It's a one-time check: "Is the condition true? If yes, do this; if no, don't do it or do something else."
- Use while when: You need to repeat an action as long as a condition remains true. It's a continuous check, a loop that keeps doing something over and over until the condition changes: "Keep doing this until the condition is no longer true."

**Summary**

- If is your go-to for one-off decisions.
- While is your choice for repeating actions based on a condition.

Think of if as a crossroads where you make a turn based on the signs, and while as a treadmill that you keep running on until you hit a stop button. Both are essential in their roles, and choosing between them depends on whether you're making a single decision or repeating actions based on ongoing conditions.

### Why Nesting is Useful

````python
current_number = 1
total_sum = 0

while current_number <= 10:
    total_sum += current_number
    current_number += 1

print(total_sum)
````

### Explanation of the Code for Activity 2: Sum of Numbers

1. **Initialization**:
   - `current_number = 1`: This initializes the starting point of the count from 1. This variable will track which number is currently being added to the sum.
   - `total_sum = 0`: This sets up an accumulator for the sum, starting at zero. This variable stores the cumulative sum of all numbers from 1 to 10 as they are added together.

2. **While Loop**:
   - The loop condition, `current_number <= 10`, ensures the loop continues to execute as long as the current number is less than or equal to 10. This condition checks if there are still numbers left to add.
   - Inside the loop, `total_sum += current_number` performs the addition of the current number to `total_sum`. This operation is critical as it accumulates the total sum of numbers.
   - `current_number += 1` increments the `current_number` by 1, moving to the next integer in the sequence. This ensures that the loop progresses by adding the next number in the sequence.

3. **Output**:
   - After the loop exits (when `current_number` exceeds 10), the program prints `total_sum`, which by now holds the sum of numbers from 1 to 10.

### Result:
The output of this script will be `55`, which represents the sum of the numbers from 1 to 10. This example demonstrates the practical application of `while` loops for performing arithmetic accumulation and controlling loops with simple numerical conditions.
